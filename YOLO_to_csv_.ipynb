```python
import os
import pandas as pd

def load_class_mapping(class_file):
    """Load class mapping from Class.txt"""
    with open(class_file, 'r', encoding='utf-8') as f:
        classes = [line.strip() for line in f.readlines()]
    return classes

def extract_classes_from_txt(txt_file, class_mapping):
    """Extract class names from annotation file"""
    extracted_classes = []
    numeric_classes = []

    with open(txt_file, 'r', encoding='utf-8') as f:
        for line in f:
            parts = line.strip().split()
            if parts and parts[0].isdigit():  # Ensure it's a valid class index
                class_idx = int(parts[0])
                if 0 <= class_idx < len(class_mapping):
                    if class_idx in range(10):  # Numeric classes (0-9)
                        numeric_classes.append((class_mapping[class_idx], float(parts[1])))  # Store class name with x-coordinate
                    else:
                        extracted_classes.append(class_mapping[class_idx])

    # Sort numeric classes by x-coordinate
    numeric_classes.sort(key=lambda x: x[1])
    sorted_numeric_class_names = [cls for cls, _ in numeric_classes]

    return ' '.join(extracted_classes + sorted_numeric_class_names)

def process_annotation_files(annotation_folder, class_file, output_csv):
    """Process all .txt annotation files and create a CSV"""
    class_mapping = load_class_mapping(class_file)
    data = []

    for file in os.listdir(annotation_folder):
        if file.endswith(".txt") and file != os.path.basename(class_file):  # Exclude Class.txt
            file_path = os.path.join(annotation_folder, file)
            class_names = extract_classes_from_txt(file_path, class_mapping)
            data.append([file, class_names])

    df = pd.DataFrame(data, columns=['File Name', 'Class Names'])
    df.to_csv(output_csv, index=False, encoding='utf-8')
    print(f"CSV file saved as {output_csv}")

# Usage example
annotation_folder = "/content/drive/MyDrive/Colab Notebooks/Dataset/Character Recognition/Bangla_License_Plate/Bangla_License_Plate"  # Change this to your directory containing .txt files
class_file = "/content/drive/MyDrive/Colab Notebooks/Dataset/Character Recognition/Bangla_License_Plate/Bangla_License_Plate/classes.txt"  # Change this to your Class.txt path
output_csv = "/content/drive/MyDrive/Colab Notebooks/Dataset/Character Recognition/Bangla_License_Plate/output1.csv"

process_annotation_files(annotation_folder, class_file, output_csv)
```

```python
import pandas as pd
import re

def extract_number(filename):
    """Extract numeric value from filename for sorting."""
    match = re.search(r'\d+', filename)
    return int(match.group()) if match else float('inf')

def sort_and_rename_csv(input_csv, output_csv):
    """Sort CSV by numerical order of filenames and replace .txt with .jpeg."""
    df = pd.read_csv(input_csv)

    # Sort by extracted numeric value from filename
    df['File Name'] = df['File Name'].astype(str)
    df = df.sort_values(by='File Name', key=lambda x: x.map(extract_number))

    # Replace .txt with .jpeg
    df['File Name'] = df['File Name'].str.replace('.txt', '.jpeg', regex=False)

    # Save to new CSV
    df.to_csv(output_csv, index=False, encoding='utf-8')
    print(f"Sorted CSV saved as {output_csv}")

# Usage example
input_csv = "/content/drive/MyDrive/Colab Notebooks/Dataset/Character Recognition/Bangla_License_Plate/output1.csv"
output_csv = "/content/drive/MyDrive/Colab Notebooks/Dataset/Character Recognition/Bangla_License_Plate/sorted_output.csv"

sort_and_rename_csv(input_csv, output_csv)
```
